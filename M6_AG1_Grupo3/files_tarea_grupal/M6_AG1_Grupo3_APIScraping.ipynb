####  Carga de datos desde la API huffpost.com
Esta rutina que hemos desarrollado como plan B ante la inestabilidad de la página, itera cada una de las categorías, haciendo las respectivas consultas a la API, datos que se van almacenando en una lista. Finalmente convertimos la lista en DataFrame y lo exportamos como CSV.   

```python
import json
import pandas as pd
from urllib.request import urlopen
sections_web = ['WELLNESS','PARENTING','TRAVEL','BUSINESS','SPORTS']
sections_api = ['healthy-living','parents','travel','business','sports']
news=[]
for i in range(0,len(sections_api)):
    url = "https://www.huffpost.com/api/section/"+sections_api[i]+"/cards?page=1&limit=20"
    response = urlopen(url)
    data_json = json.loads(response.read())
    news.extend([{'category':sections_web[i], 'headline':e['headlines'][0]['text']} for e in data_json['cards']])
df= pd.DataFrame(news)
df
df.to_csv('huffpost.csv',index=False)
```
